# Q5.

````python
home_games_by_team = df.groupby('home_team')['game_id'].nunique()
team_with_most_home_games = home_games_by_team.idxmax()
num_home_games = home_games_by_team.max()

print(f"O time com mais jogos em casa é o {team_with_most_home_games} com {num_home_games} jogos.")
````

# Q6.
Qual foi o percentual de partidas ganhas pelo "Bufalo Bulls" considerados todos os jogos da base em que participou?

````python
buffalo_bulls_games = df[df['team'] == 'BUF']

wins = 0
losses = 0

for index, row in buffalo_bulls_games.iterrows():
  if row['home_team'] == 'BUF':
    if row['home_score'] > row['vis_score']:
      wins += 1
    else:
      losses += 1
  else:
    if row['vis_score'] > row['home_score']:
      wins += 1
    else:
      losses += 1

total_games = wins + losses

win_percentage = (wins / total_games) * 100
print(f"O percentual de partidas ganhas pelo Buffalo Bulls é de {win_percentage:.2f}%")
````

# Q8.
(NFL) Você pode afirmar, com base nos jogos da base, que os resultados dos jogos (vitórias) favorecem o time "da casa"?

````python
home_team_wins = 0
away_team_wins = 0

for index, row in df.iterrows():
  if row['home_score'] > row['vis_score']:
    home_team_wins += 1
  elif row['vis_score'] > row['home_score']:
    away_team_wins += 1

total_games = home_team_wins + away_team_wins

if total_games > 0:
  home_team_win_percentage = (home_team_wins / total_games) * 100
  print(f"Percentual de vitórias do time da casa: {home_team_win_percentage:.2f}%")

  if home_team_win_percentage > 50:
    print("Com base nos jogos da base, os resultados favorecem o time da casa.")
  else:
    print("Com base nos jogos da base, os resultados não favorecem o time da casa.")
````

# Q9.
Considerando todo o período, qual a valorização dos ativos de maior e menor ganho?

````python
df2 = pd.read_csv('https://github.com/Rogerio-mack/IMT_CD_2025/raw/refs/heads/main/data/stocks_2019_2025.csv')
df2.head()
````

````python
ativos = ["BRL=X", "BTC-USD", "ITUB3.SA", "PETR4.SA", "VALE3.SA", "^BVSP"]

valorizacoes = {}

for ativo in ativos:
    df_ativo = df2[["Date", ativo]].dropna().sort_values("Date")
    if not df_ativo.empty:
        primeiro_valor = df_ativo[ativo].iloc[0]
        ultimo_valor = df_ativo[ativo].iloc[-1]
        valorizacao = ((ultimo_valor - primeiro_valor) / primeiro_valor) * 100
        valorizacoes[ativo] = valorizacao

ativo_max = max(valorizacoes, key=valorizacoes.get)
ativo_min = min(valorizacoes, key=valorizacoes.get)

print(f"Ativo com maior valorização: {ativo_max} ({valorizacoes[ativo_max]:.2f}%)")
print(f"Ativo com menor valorização: {ativo_min} ({valorizacoes[ativo_min]:.2f}%)")
````

````python
first_values = df2.iloc[0, 1:7]
last_values = df2.iloc[-1, 1:7]

valorization = (last_values / first_values) - 1

max_valorization = valorization.max()
min_valorization = valorization.min()

max_valorization_asset = valorization[valorization == max_valorization].index[0]
min_valorization_asset = valorization[valorization == min_valorization].index[0]

print(f"O ativo de maior valorização foi {max_valorization_asset} com {max_valorization:.2%}")
print(f"O ativo de menor valorização foi {min_valorization_asset} com {min_valorization:.2%}")
````

# Q10.
Uma métrica comum na avaliação de ativos é verificarmos, em um período, a quantidade de dias em que o ativo teve valorização com relação ao dia anterior (**up and down**). Considerando todo o período, qual o percentual de `ups` dos ativos com mais e menos `ups`.

````python
ativos = ["BRL=X", "BTC-USD", "ITUB3.SA", "PETR4.SA", "VALE3.SA", "^BVSP"]

valorizacoes = {}
ups_percentual = {}

for ativo in ativos:
    df_ativo = df2[["Date", ativo]].dropna().sort_values("Date")
    if not df_ativo.empty:
        primeiro_valor = df_ativo[ativo].iloc[0]
        ultimo_valor = df_ativo[ativo].iloc[-1]
        valorizacao = ((ultimo_valor - primeiro_valor) / primeiro_valor) * 100
        valorizacoes[ativo] = valorizacao

        df_ativo["variação"] = df_ativo[ativo].diff()
        ups = (df_ativo["variação"] > 0).sum()
        total_dias = df_ativo["variação"].count()
        percentual_ups = (ups / total_dias) * 100 if total_dias > 0 else 0
        ups_percentual[ativo] = percentual_ups

ativo_max = max(valorizacoes, key=valorizacoes.get)
ativo_min = min(valorizacoes, key=valorizacoes.get)

ativo_mais_ups = max(ups_percentual, key=ups_percentual.get)
ativo_menos_ups = min(ups_percentual, key=ups_percentual.get)

print(f"Ativo com maior valorização: {ativo_max} ({valorizacoes[ativo_max]:.2f}%)")
print(f"Ativo com menor valorização: {ativo_min} ({valorizacoes[ativo_min]:.2f}%)")
print(f"Ativo com maior percentual de ups: {ativo_mais_ups} ({ups_percentual[ativo_mais_ups]:.2f}%)")
print(f"Ativo com menor percentual de ups: {ativo_menos_ups} ({ups_percentual[ativo_menos_ups]:.2f}%)")
````

# Q12.
A *MAIOR PARTE* dos ativos da base teve sua maior valorização em um único dia no mesmo ano e mês. Qual foi esse ano-mês especial?

````python
ativos = ["BRL=X", "BTC-USD", "ITUB3.SA", "PETR4.SA", "VALE3.SA", "^BVSP"]

valorizacoes = {}
ups_percentual = {}

for ativo in ativos:
    df_ativo = df2[["Date", ativo]].dropna().sort_values("Date")
    if not df_ativo.empty:
        primeiro_valor = df_ativo[ativo].iloc[0]
        ultimo_valor = df_ativo[ativo].iloc[-1]
        valorizacao = ((ultimo_valor - primeiro_valor) / primeiro_valor) * 100
        valorizacoes[ativo] = valorizacao

        df_ativo["variação"] = df_ativo[ativo].diff()
        ups = (df_ativo["variação"] > 0).sum()
        total_dias = df_ativo["variação"].count()
        percentual_ups = (ups / total_dias) * 100 if total_dias > 0 else 0
        ups_percentual[ativo] = percentual_ups

ativo_max = max(valorizacoes, key=valorizacoes.get)
ativo_min = min(valorizacoes, key=valorizacoes.get)

ativo_mais_ups = max(ups_percentual, key=ups_percentual.get)
ativo_menos_ups = min(ups_percentual, key=ups_percentual.get)

df2["Ano-Mês"] = df2["Date"].astype(str).str[:7]
maiores_valorizacoes = {}

for ativo in ativos:
    if f"{ativo} %" in df2.columns:
        df_max = df2.loc[df2[f"{ativo} %"].idxmax()]
        maiores_valorizacoes[ativo] = df_max["Ano-Mês"]

ano_mes_especial = max(set(maiores_valorizacoes.values()), key=list(maiores_valorizacoes.values()).count)

print(f"Ativo com maior valorização: {ativo_max} ({valorizacoes[ativo_max]:.2f}%)")
print(f"Ativo com menor valorização: {ativo_min} ({valorizacoes[ativo_min]:.2f}%)")
print(f"Ativo com maior percentual de ups: {ativo_mais_ups} ({ups_percentual[ativo_mais_ups]:.2f}%)")
print(f"Ativo com menor percentual de ups: {ativo_menos_ups} ({ups_percentual[ativo_menos_ups]:.2f}%)")
print(f"Ano-mês com maior valorização para a maioria dos ativos: {ano_mes_especial}")
````

# Q13.
Qual a média de variação do dólar no período, excluídos os *outliers* (atenção: não é necessário fazer qualquer tratamento para os valores ausentes).

````python
ativos = ["BRL=X", "BTC-USD", "ITUB3.SA", "PETR4.SA", "VALE3.SA", "^BVSP"]

valorizacoes = {}
ups_percentual = {}

for ativo in ativos:
    df_ativo = df2[["Date", ativo]].dropna().sort_values("Date")
    if not df_ativo.empty:
        primeiro_valor = df_ativo[ativo].iloc[0]
        ultimo_valor = df_ativo[ativo].iloc[-1]
        valorizacao = ((ultimo_valor - primeiro_valor) / primeiro_valor) * 100
        valorizacoes[ativo] = valorizacao

        df_ativo["variação"] = df_ativo[ativo].diff()
        ups = (df_ativo["variação"] > 0).sum()
        total_dias = df_ativo["variação"].count()
        percentual_ups = (ups / total_dias) * 100 if total_dias > 0 else 0
        ups_percentual[ativo] = percentual_ups

ativo_max = max(valorizacoes, key=valorizacoes.get)
ativo_min = min(valorizacoes, key=valorizacoes.get)

ativo_mais_ups = max(ups_percentual, key=ups_percentual.get)
ativo_menos_ups = min(ups_percentual, key=ups_percentual.get)

df2["Ano-Mês"] = df2["Date"].astype(str).str[:7]
maiores_valorizacoes = {}

for ativo in ativos:
    if f"{ativo} %" in df2.columns:
        df_max = df2.loc[df2[f"{ativo} %"].idxmax()]
        maiores_valorizacoes[ativo] = df_max["Ano-Mês"]

ano_mes_especial = max(set(maiores_valorizacoes.values()), key=list(maiores_valorizacoes.values()).count)

if "BRL=X %" in df2.columns:
    variacao_dolar = df2["BRL=X %"].dropna()
    q1 = variacao_dolar.quantile(0.25)
    q3 = variacao_dolar.quantile(0.75)
    iqr = q3 - q1
    limite_inferior = q1 - 1.5 * iqr
    limite_superior = q3 + 1.5 * iqr
    variacao_filtrada = variacao_dolar[(variacao_dolar >= limite_inferior) & (variacao_dolar <= limite_superior)]
    media_variacao_dolar = variacao_filtrada.mean()
else:
    media_variacao_dolar = None

print(f"Ativo com maior valorização: {ativo_max} ({valorizacoes[ativo_max]:.2f}%)")
print(f"Ativo com menor valorização: {ativo_min} ({valorizacoes[ativo_min]:.2f}%)")
print(f"Ativo com maior percentual de ups: {ativo_mais_ups} ({ups_percentual[ativo_mais_ups]:.2f}%)")
print(f"Ativo com menor percentual de ups: {ativo_menos_ups} ({ups_percentual[ativo_menos_ups]:.2f}%)")
print(f"Ano-mês com maior valorização para a maioria dos ativos: {ano_mes_especial}")
print(f"Média da variação do dólar excluindo outliers: {media_variacao_dolar:.2f}%")
````

# Q15.

(European Energy) Qual percentual de energia nuclear produzida na Europa no ano de 2018?

````python
nuclear_energy_2018 = df3[(df3['type'] == 'Nuclear') & (df3['2018'].notnull())]['2018'].sum()
total_energy_2018 = df3[df3['2018'].notnull()]['2018'].sum()

if total_energy_2018 > 0:
  percentage_nuclear_2018 = (nuclear_energy_2018 / total_energy_2018) * 100
  print(f"O percentual de energia nuclear produzida na Europa em 2018 foi de {percentage_nuclear_2018:.2f}%")
````
