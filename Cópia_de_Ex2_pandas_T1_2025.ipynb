# **CASE: Gapminder**

# Q1.

Considere os dados de `coal_eletricity`. Reorganize os dados como abaixo selecionando somente dados de 2011 em diante. Salve o resultado em um dataframe `coal_horiz`. Faça uma solução sem laços de programa.



````python
coal_horiz = coal_eletricity[coal_eletricity['year'] >= 2011].pivot(index='geo', columns='year', values='coal_electricity_generation_twh')
coal_horiz.head()
````

# Q2.

Considere os dados de `coal_eletricity`. Crie para cada localização geográfica (`geo`) os valores 'all' e 'mean' com a soma e média de valores de todos os anos como abaixo. Salve o resultado em um dataframe `coal_verti`. Elabore uma solução sem laços de programa.


````python
df_agg = coal_eletricity.groupby('geo')['coal_electricity_generation_twh'].agg(all='sum', mean='mean').reset_index()

coal_verti = df_agg.melt(id_vars='geo', var_name='year', value_name='coal_electricity_generation_twh')

coal_verti = pd.concat([coal_eletricity , coal_verti], ignore_index=True).sort_values(by=['geo', 'year'], ascending=[True, True])
coal_verti
````

````python
#@markdown sanity check
print(coal_verti.shape == (1628, 3))
````

````python
coal_verti.loc[random.sample(range(len(coal_verti)), 10)]
````

Responda em Q2 com o valor da soma abaixo em:

````python
random.seed(1)
int(coal_verti.loc[random.sample(range(len(coal_verti)), 10)].coal_electricity_generation_twh.sum())
````

# Q3.

Considere os dados de `coal_eletricity`. Determine qual país teve a maior produção de energia elétrica por carvão a cada ano. Note que os dados possuem valores totais por região e podem precisar serem excluídos. Elabore uma solução sem laços de programa.

````python
excluded_geos = ['total_world', 'total_north_america', 'total_asia_pacific', 'total_europe', 'total_africa', 'total_cis']

df_filtered = coal_eletricity[~coal_eletricity['geo'].isin(excluded_geos)]

idx_max_por_ano = df_filtered.groupby('year')['coal_electricity_generation_twh'].idxmax()
df_max_por_ano = df_filtered.loc[idx_max_por_ano].reset_index(drop=True)
df_max_por_ano
````

# Q4. Pareto 80/20

Considere os dados de `coal_eletricity`. **Encontre os maiores geradores em 2019 e determine o pequeno número de países responsáveis por 80% da geração mundial**. Note que os dados possuem valores totais por região que precisam ser excluídos. Você é livre para elaborar uma solução com ou sem laços de programa (sim, sem laço é possível).

**Dica.** Você pode querer empregar o comando [`cumsum`](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.cumsum.html)


````python
df_2019 = df_filtered[df_filtered['year'] == 2019]

df_2019 = df_2019.groupby('geo', as_index=False)['coal_electricity_generation_twh'].sum()

df_2019 = df_2019.sort_values('coal_electricity_generation_twh', ascending=False)

total_2019 = df_2019['coal_electricity_generation_twh'].sum()
df_2019['cumulative_sum'] = df_2019['coal_electricity_generation_twh'].cumsum()
df_2019['cumulative_pct'] = df_2019['cumulative_sum'] / total_2019

df_2019_pareto = df_2019[df_2019['cumulative_pct'] <= 0.85]

df_2019_pareto
````

Nesta questão responda sobre os países maiores geradores de energia que correspondem a 80% da produção mundial dos países.

# Q5. Como se relacionam as emissões de CO2 e a geração de energia por carvão?

Veja o exemplo de join a seguir. E você pode querer acessar a documentação do `Pandas` para uso do [`merge`](https://pandas.pydata.org/docs/reference/api/pandas.merge.html).





````python
country_entities = pd.read_csv('https://github.com/open-numbers/ddf--gapminder--co2_emission/raw/master/ddf--entities--country.csv')
country_entities.head()
````

````python
co2_emissions = pd.read_csv('https://github.com/open-numbers/ddf--gapminder--co2_emission/raw/master/ddf--datapoints--yearly_co2_emissions_1000_tonnes--by--country--year.csv')
co2_emissions.head()
````

````python
co2_country = country_entities.merge(co2_emissions, how='inner', left_on='country', right_on='country')

# ou simplesmente,
# co2_country = country_entities.merge(co2_emissions)
# já que colunas de mesmo nome são automaticamente empregadas no join

co2_country
````

Empregue o exemplo anterior e o dataframe `coal_eletricity` para construir um conjunto de dados `co2_coal` como abaixo.

Note que os nomes dos países encontram-se em um formato diferente (lower/upper case) nos dataframes a serem empregados e você pode querer empregar a função [`str`](https://pandas.pydata.org/docs/reference/api/pandas.Series.str.html) para ajustar isso. Você pode empregar somente os nomes que satisfizerem a junção interna.

````python
country_entities['name'] = country_entities['name'].str.lower()
coal_eletricity['geo'] = coal_eletricity['geo'].str.lower()
co2_coal = coal_eletricity.merge(country_entities, how='inner', right_on='name', left_on='geo').merge(co2_emissions).drop(columns=['geo'])
co2_coal.head()
````

Responda Q5 com a correlação abaixo.





````python
np.round(co2_coal[['yearly_co2_emissions_1000_tonnes','coal_electricity_generation_twh']].corr().values[0,1],5)
````

# Q6.
