# **Case**: Airbnb

Neste case, analise dados sobre imóveis alugados pela plataforma Airbnb em NYC no ano de 2019.


```python
import pandas as pd
import numpy as np

df = pd.read_csv(path + 'AB_NYC_2019.csv')
df
```

# Q1. Proprietários com mais imóveis na plataforma

Obtenha os 3 proprietários ("hosts") com mais imóveis sendo anunciados na plataforma, e a respectiva quantidade de imóveis anunciados.


```python
host_counts = df.groupby('host_id')['id'].count().reset_index()
host_counts.columns = ['host_id', 'number_of_listings']
host_counts = host_counts.sort_values('number_of_listings', ascending=False)
top_hosts = pd.merge(host_counts.head(3), df[['host_id', 'host_name']], on='host_id', how='left')
top_hosts = top_hosts.drop_duplicates(subset=['host_id', 'host_name'])

print(top_hosts[['host_name', 'number_of_listings']])
```

# Q2B. Preço nos bairros com mais imóveis

 Qual o preço médio dos imóveis, considerando somente os preços dos imóveis cujos bairros (`neighbourhood`) tenham mais de 20 imóveis sendo anunciados?  


```python
bairros_contagem = df['neighbourhood'].value_counts()
bairros_validos = bairros_contagem[bairros_contagem > 20].index
df_filtrado = df[df['neighbourhood'].isin(bairros_validos)]
preco_medio = df_filtrado['price'].mean()

print(f'O preço médio dos imóveis em bairros com mais de 20 anúncios é: USD {preco_medio:.2f}')
```

# Q3B.  Brooklyn *outliers*

Qual o percentual de imóveis de Manhattan com preços estatisticamente discrepantes (*outliers*) com relação aos preços de todos os imóveis?



```python
brooklyn = df[df['neighbourhood_group'] == 'Brooklyn']

q1 = brooklyn['price'].quantile(0.25)
q3 = brooklyn['price'].quantile(0.75)
iqr = q3 - q1

limite_inferior = q1 - 1.5 * iqr
limite_superior = q3 + 1.5 * iqr

outliers = brooklyn[(brooklyn['price'] < limite_inferior) | (brooklyn['price'] > limite_superior)]

perc_brooklyn = (len(outliers) / len(brooklyn)) * 100

print(f'Percentuais de outliers de preços, Brooklyn: {perc_brooklyn:.2f} %')
```

# Q4B. Scores

Nesta tarefa você deve atribuir a cada imóvel um valor de *score* (`df['score']`) baseado nas seguintes regras:

1. Imóveis com `availability_365` > 90 recebem +10 pontos enquanto imóveis com `availability_365` < 5 recebem -5 pontos;
2. Imóveis com o número de reviews >= 10 recebem +5 pontos para cada 10 reviews (por ex. se número de reviews = 20, são atribuídos +10 pontos). Se número de reviews = 0 o imóvel recebe -5 pontos;
3. Imóveis com preço abaixo da média de seu `neighbourhood`	recebem +5 pontos.

Implemente empregando `apply()` ou um laço de programa, como preferir.

```python
preco_medio_bairro = df.groupby('neighbourhood')['price'].mean()

def calcular_score(x):
    score = 0

    if x['availability_365'] > 90:
        score += 10
    elif x['availability_365'] < 5:
        score -= 5

    if x['number_of_reviews'] == 0:
        score -= 5
    else:
        score += (x['number_of_reviews'] // 10) * 5

    media_bairro = preco_medio_bairro.get(x['neighbourhood'], None)
    if media_bairro and x['price'] < media_bairro:
        score += 5

    return score

df['score'] = df.apply(calcular_score, axis=1)

print(f'A soma dos valores de score é: ', df['score'].sum())
```

# Q5. Enriquecendo os dados

Enriqueça os dados dos imóveis com os valores dos impostos e multas municipais. Cacule a média de impostos e multas de cada grupo de vizinhaça, `neighbourhood_group`. Note que existem imóveis sem impostos e sem multas, esses são imóveis isentos, mas devem ser considerados nos cálculos de média (por exemplo, você pode considerar esses imóveis com valor 0).

```python
df_taxes = pd.read_csv(path + 'AB_NYC_2019_taxes.csv')
df_taxes
```

```python
df_taxes_pivotado = df_taxes.pivot_table(index='id', columns='type', values='value', aggfunc='sum')
df_taxes_pivotado = df_taxes_pivotado.reset_index()
df_taxes_pivotado = df_taxes_pivotado.fillna(0)

df_completo = df.merge(df_taxes_pivotado, how='left', on='id')
df_completo[['fines', 'taxes']] = df_completo[['fines', 'taxes']].fillna(0)
medias = df_completo.groupby('neighbourhood_group')[['fines', 'taxes']].mean()

print("Média de multas e impostos por neighbourhood_group:")
print(medias.round(2))
```

# Q6. Quebras

Construa no `Pandas` um "relatório" (`df_report`) como abaixo. Os valores são ordenados `neighbourhood_group`,`neighbourhood`. Cada "página" contém apenas 30 linhas. A cada "quebra" de `neighbourhood_group` é gerada uma linha com preço médio do `neighbourhood_group` e, então, é iniciada uma nova página.

O item `line_id` é opcional mas deve auxiliá-lo nessa tarefa.

Ao final produza a soma dos preços das páginas abaixo...

```python
sum_price_pages = df_report[ df_report.page_id.isin([1,24,25,26,28]) ].groupby('page_id').price.sum().astype(int).sum()

print(f'A soma dos preços das páginas 1, 24, 25 e 28 é: {sum_price_pages}')
```

```python
df_sorted = df.sort_values(['neighbourhood_group', 'neighbourhood']).reset_index(drop=True)

linhas = []

line_id = 0
page_id = 1
linhas_na_pagina = 0
grupo_anterior = None

for _, row in df_sorted.iterrows():
    grupo_atual = row['neighbourhood_group']

    if grupo_atual != grupo_anterior and grupo_anterior is not None:
        media_preco = df_sorted[df_sorted['neighbourhood_group'] == grupo_anterior]['price'].mean()
        linhas.append({
            'neighbourhood_group': grupo_anterior,
            'neighbourhood': 'MEAN',
            'price': media_preco,
            'page_id': page_id,
            'line_id': line_id
        })
        line_id += 1
        linhas_na_pagina += 1

        if linhas_na_pagina >= 30:
            page_id += 1
            linhas_na_pagina = 0
        else:
            page_id += 1
            linhas_na_pagina = 0

    linhas.append({
        'line_id': line_id,
        'page_id': page_id,
        'neighbourhood_group': grupo_atual,
        'neighbourhood': row['neighbourhood'],
        'price': row['price']
    })
    line_id += 1
    linhas_na_pagina += 1

    if linhas_na_pagina >= 30:
        page_id += 1
        linhas_na_pagina = 0

    grupo_anterior = grupo_atual

if grupo_anterior is not None:
    media_preco = df_sorted[df_sorted['neighbourhood_group'] == grupo_anterior]['price'].mean()
    linhas.append({
        'line_id': line_id,
        'page_id': page_id,
        'neighbourhood_group': grupo_anterior,
        'neighbourhood': 'MEAN',
        'price': media_preco
    })

df_report = pd.DataFrame(linhas)

sum_price_pages = df_report[df_report.page_id.isin([1,24,25,26,28])].groupby('page_id')['price'].sum().astype(int).sum()

print(f'A soma dos preços das páginas 1, 24, 25 e 28 é: {sum_price_pages}')
```
